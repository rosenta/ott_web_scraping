```python
def start_page_items(
        tray_id: str,
        tray_type: str,
):
    
    content_vertical_img_urls = {}

    json_data = {
        'deeplink_url': f'/in/browse/{tray_type}?tray_id={tray_id}',
        'app_launch_count': 100,
    }

    response = requests.post('https://www.hotstar.com/api/internal/bff/v2/start', 
                            headers=headers, 
                            json=json_data
                            )

    if response.status_code == 200:
        response_data = response.json()
        if response_data.get('success'):
            for ix, widget_data in enumerate(response_data["success"]["page"]["spaces"]["content"]["widget_wrappers"][0]["widget"]["data"]["items"]):
                content_poster_data = widget_data["vertical_content_poster"]["data"]
                image_suffix = content_poster_data["image"]["src"]
                content_title = content_poster_data["expanded_content_poster"]["content_info"]["title"]
                image_url = f"https://img10.hotstar.com/image/upload/{image_suffix}"
                # print(f"{ix+1} -> {content_title} -> {image_url}")
                content_vertical_img_urls.update({
                    content_title: image_url,
                })
        else:
            print("***** Data retrieval was not successful! *****")

    return content_vertical_img_urls
    
```

```python
def get_tokenised_items(
        tray_id: str,
        tray_type: str,
        offset_start: int,
        offset_limit: int,
        token_data: dict = {},
        ipv_mode: str = "sort",
        ):
    
    content_vertical_img_urls = {}

    if not token_data:
        token_data = {
            "offset": offset_start,
            "limit": offset_limit,
            "collectionId": tray_id,
        }
        token_data = json.dumps(token_data)


    params = {
        'token': token_data,
        'tray_id': tray_id,
        'tray_type': tray_type,
        "ipv_mode": ipv_mode
    }

    response = requests.get(
        'https://www.hotstar.com/api/internal/bff/v2/pages/1049/spaces/1776/widgets/6803/items',
        params=params,
        headers=headers,
    )
    try:
        if response.status_code == 200:
            response_data = response.json()
            for ix, widget_data in enumerate(response_data["success"]["widget_wrapper"]["widget"]["data"]["items"]):
                content_poster_data = widget_data["vertical_content_poster"]["data"]
                image_suffix = content_poster_data["image"]["src"]
                content_title = content_poster_data["expanded_content_poster"]["content_info"]["title"]
                image_url = f"https://img10.hotstar.com/image/upload/{image_suffix}"
                # print(f"{ix+1} -> {content_title} -> {image_url}")
                content_vertical_img_urls.update({
                    content_title: image_url,
                })
    except Exception as e:
        print(response_data)
        print(f"params: {params}")
        raise e

    return content_vertical_img_urls
```

```python

tray_config_data_list = json.load(open(config_file_path))
```

```python
buckets_contents = {}

for tray_config in tray_config_data_list:
    tray_title = tray_config["tray_title"]
    tray_id = tray_config["tray_id"]
    tray_type = tray_config["tray_type"]
    offset_iteration = tray_config.get("offset_iteration", 1)
    max_offset_limit = tray_config.get("offset_limit", None)
    tokens = tray_config.get("tokens", [])


    print(f"\nProcessing Data for Bucket: {tray_title}")
    # print(f"-"*120)

    content_vertical_img_urls = {}
    if max_offset_limit != None:

        start = 0
        for ix in range(offset_iteration):     
            end = len(content_vertical_img_urls) + int(max_offset_limit/(offset_iteration))
            print(f"*************** ix {ix+1}: offset from {start} to {end}")
            content_vertical_img_urls.update(
                get_tokenised_items(
                    tray_id=tray_id,
                    tray_type=tray_type,
                    offset_start=start,
                    offset_limit=end,
                )
            )
            start = end

    else:
        content_vertical_img_urls.update(
            start_page_items(
                tray_id=tray_id,
                tray_type=tray_type,
            )
        )

        for token in tokens:
            content_vertical_img_urls.update(
                get_tokenised_items(
                    tray_id=tray_id,
                    tray_type=tray_type,
                    token_data=token,
                    offset_start=0,
                    offset_limit=10,                    
                )
            )

    print(f"Count of images for bucket:'{tray_title}' -> {len(content_vertical_img_urls)}")
    buckets_contents.update({
        tray_title: content_vertical_img_urls
    })
    print("#"*150)
```

```python
with open("buckets_contents_image_urls.json", "w+", encoding="utf-8") as f:
    json.dump(buckets_contents, f)
```
